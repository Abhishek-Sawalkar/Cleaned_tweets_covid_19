```python
import pandas as pd

df = pd.read_csv("NewMarch.csv")
```

```python
import re
from datetime import datetime

def date_extract(text):
    
    match = re.search(r'\d{4}-\d{2}-\d{2}', text)
    date = datetime.strptime(match.group(), '%Y-%m-%d').date()
    return date


df["Datetime"] = df["Datetime"].apply(date_extract)
df["Datetime"].head()
```

```python
from statistics import mean

def ok(da):
    temp = df["Polarity"].where(df['Datetime']==t[0])
    temp = [x for x in temp if str(x) != 'nan']
    lul = mean(temp)
    print(lul)
    return (da, lul)

    
lst = []
for da in t:
    print()
    print(f'date:{da}')
    lst.append(ok(da))
```

```python
dff = pd.DataFrame(lst)
dff.columns = ['date', 'Polarity']
```
